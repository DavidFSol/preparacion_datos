```python
data_export = data_export.merge(data_gdp[["country", "Country Code"]], how='outer')
data_import = data_import.merge(data_gdp[["country", "Country Code"]], how='outer')
data_export.head()
```

```python
data_ex_imp = data_export.merge(data_import, how='outer', on=['HSCode', 'country', 'year'])
data_ex_imp
```

```python
for k, v in enumerate(data_ex_imp.values):
    if pd.isna(data_ex_imp.loc[k, 'Commodity_x']) and not pd.isna(data_ex_imp.loc[k, 'Commodity_y']):
        data_ex_imp.loc[k, 'Commodity_x'] = data_ex_imp.loc[k, 'Commodity_y']
    if pd.isna(data_ex_imp.loc[k, 'Country Code_x']) and not pd.isna(data_ex_imp.loc[k, 'Country Code_y']):
        data_ex_imp.loc[k, 'Country Code_x'] = data_ex_imp.loc[k, 'Country Code_y']

data_ex_imp = data_ex_imp[['HSCode', 'Commodity_x', 'country', 'Country Code_x', 'year', 'import value', 'export value']]
data_ex_imp
```

```python
data_ex_imp.rename(columns = {'Commodity_x':'Commodity', 'Country Code_x':'Country Code'}, inplace = True)
data_ex_imp
```

```python
data_gdp.set_index('Country Code', inplace=True)
data_gdp.head()
```

```python
data_ex_imp = data_ex_imp.dropna(subset=['HSCode', 'Commodity', 'country', 'Country Code', 'year'])
data_ex_imp[['HSCode', 'year']] = data_ex_imp[['HSCode', 'year']].astype(int)
data_ex_imp['gdp year'] = [data_gdp.loc[v[-4], str(v[-3])] for k, v in enumerate(data_ex_imp.values)]
data_ex_imp.head()
```

```python
data_ex_imp_income = data_ex_imp.merge(data_income, on='Country Code', how='outer')
data_ex_imp_income.head()
```

```python
data_pop.rename(columns = {'Country Name':'country', 'Year':'year', 'Count':'Pop.Count'}, inplace = True)
data_pop.head()
```

```python
data_ex_imp_pop = data_ex_imp_income.merge(data_pop, how='left', on=['country', 'year'])
data_ex_imp_pop = data_ex_imp_pop.dropna(subset=['HSCode', 'Commodity', 'country', 'Country Code', 'year', 'gdp year', 'Pop.Count']).reset_index()
data_ex_imp_pop[['HSCode', 'year']] = data_ex_imp_pop[['HSCode', 'year']].astype(int)
data_ex_imp_pop
```

```python
rel = data_relig.iloc[:, 3:11]
data_relig['Main Religion'] = np.nan * data_relig.shape[0]

data_relig['Main Religion'] = [rel.loc[k].idxmax() for k in rel.index]
data_relig
```

```python
data_relig_short = data_relig[['Country', 'Main Religion']].drop_duplicates(subset=['Country'], keep='first').set_index('Country')
data_relig_short
```

```python
df = data_ex_imp_pop
df['Religion'] = np.nan * data_ex_imp_pop.shape[0]

country_list = data_relig_short.index.values.tolist()
i = 0

while i < df.shape[0]:
    country = df.loc[i, 'country']
    j = df[df['country'] == country].index[-1]
    if country in country_list:
        df['Religion'][df['country'] == country] = data_relig_short.loc[country][0]
        i = j + 1
    else: i += 1

df
```
